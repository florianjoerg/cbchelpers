# Examples on how to use cbchelpers

## First we will start by exploring the available colors

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
from cbchelpers import UniColors as uc
import matplotlib.pyplot as plt

# get all the available colors
uc.show(short=True)

# use a specific color in a plot
x = np.arange(100)
y = np.random.randn(100)
plt.title("First Plot")
plot = plt.plot(x, y, color=uc.BLUE.value)
plt.show()
plt.close()
# they are also availabel as context manager
from cbchelpers import UniColorsContext
with UniColorsContext():
    plt.title("PLot with context manager")
    for i in range(5):
        plt.plot(x,y+5*i)
plt.show()
plt.close()

plt.title("Plot with matplotlib colors")
for i in range(5):
    plt.plot(x,y+5*i)
plt.show()
plt.close()

# they can also be set as default instead of the matplotlib colors
uc.set_as_default()
plt.title("Plot with UniColors as default")
for i in range(5):
    plt.plot(x,y+5*i)
plt.show()
plt.close()

#there are also registered colormaps:
# ...

plt.rcdefaults()

plt.title("Plot with matplotlib colors")
for i in range(5):
    plt.plot(x,y+5*i)
plt.show()
plt.close()
```
